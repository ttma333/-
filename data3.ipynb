```python
import os
import pandas as pd

folder_path = "C:/Users/ttma9/downloads/Brawlstars_project-main-20230313T042453Z-001/Brawlstars_project-main"


csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]


dfs = []

for file in csv_files:
    df = pd.read_csv(os.path.join(folder_path, file))
    dfs.append(df)

merged_df = pd.concat(dfs, ignore_index=True)

merged_df = merged_df.drop(merged_df.columns[0], axis=1)

merged_df
```

```python
df = merged_df.copy()
```

```python
df=df[~df['result'].isna()]
df
```

```python
df.drop(columns=['rank','brawler_id'],axis=1,inplace=True)
df
```

```python
def get_result(row):
    if row['team'] == 1:
        if row['result'] == 'victory':
            return 'win'
        else:
            return 'defeat'
    else:
        if row['result'] == 'victory':
            return 'defeat'
        else:
            return 'win'

df['new_result'] = df.apply(get_result, axis=1)
df
```

```python
player_map_win_rates = df.groupby(['player_tag', 'map'])['new_result'].apply(lambda x: (x=='win').sum() / len(x)).reset_index()
player_map_win_rates.rename(columns={'new_result': 'win_rate'}, inplace=True)

# Print results 
print("player_tag 따른 맵별 승률:")
player_map_win_rates
```

```python
brawler_mode_win_rates = df.groupby(['brawler_name', 'mode'])['new_result'].apply(lambda x: (x=='win').sum() / len(x)).reset_index()
brawler_mode_win_rates.rename(columns={'new_result': 'win_rate'}, inplace=True)

print("브롤러에따른 모드별 승률:")
brawler_mode_win_rates
```

```python
brawler_map_win_rates = df.groupby(['brawler_name', 'map'])['new_result'].apply(lambda x: (x=='win').sum() / len(x)).reset_index()
brawler_map_win_rates.rename(columns={'new_result': 'win_rate'}, inplace=True)

print("브롤러에따른 모드별 승률:")
brawler_map_win_rates
```

```python
brawler_mode_win_rates.to_csv('mode_winrate.csv')
```

```python
brawler_map_win_rates.to_csv('map_winrate.csv')
```
